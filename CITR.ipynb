```python
# ===========================
# CITR (Cleaning, Integration, Transformation, Reduction)
# ===========================

import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Load Dataset
df = pd.read_csv("train.csv")

print("Original Shape:", df.shape)

# ===========================
# C - DATA CLEANING
# ===========================

# Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df["Cabin"] = df["Cabin"].fillna("Unknown")

# Remove duplicate rows
df = df.drop_duplicates()

print("\nCleaning Completed")
print("Shape after Cleaning:", df.shape)

# ===========================
# I - DATA INTEGRATION
# ===========================

# Example integration (self merge)
extra = df[["PassengerId", "Fare"]]
df = pd.merge(df, extra, on="PassengerId", suffixes=("", "_copy"))

print("\nIntegration Completed")
print("Shape after Integration:", df.shape)

# ===========================
# T - DATA TRANSFORMATION
# ===========================

# Encode categorical columns
le = LabelEncoder()

df["Sex"] = le.fit_transform(df["Sex"])
df["Embarked"] = le.fit_transform(df["Embarked"])
df["Cabin"] = le.fit_transform(df["Cabin"])

# Normalize numerical columns
scaler = MinMaxScaler()

df[["Age", "Fare"]] = scaler.fit_transform(df[["Age", "Fare"]])

print("\nTransformation Completed")

# ===========================
# R - DATA REDUCTION
# ===========================

# Remove unnecessary columns
df = df.drop(["PassengerId", "Name", "Ticket", "Cabin", "Fare_copy"], axis=1)

print("\nReduction Completed")
print("Final Shape:", df.shape)

# ===========================
# DISPLAY FINAL OUTPUT
# ===========================

print("\nFinal Dataset:")
print(df.head())

# Save final dataset
df.to_csv("reduced_train.csv", index=False)

print("\nFinal dataset saved as 'reduced_train.csv'")
```
